```python
# Instalación de librerías necesarias
!pip install pandas matplotlib seaborn -q

# Importación de librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("✅ Librerías cargadas correctamente")
```

```python
# Carga del dataset público de salud (Heart Disease - UCI)
url = "https://raw.githubusercontent.com/sharmaroshan/Heart-UCI-Dataset/master/heart.csv"

df = pd.read_csv(url)

# Vista inicial del dataset
print("✅ Dataset cargado correctamente")
print(f"📊 Filas: {df.shape[0]} | Columnas: {df.shape[1]}")
print("\n--- Primeras 5 filas ---")
df.head()
```

```python
# Renombrar columnas al español para mejor comprensión
df.columns = [
    'edad', 'sexo', 'tipo_dolor_pecho', 'presion_arterial',
    'colesterol', 'azucar_ayunas', 'ecg_reposo', 'frec_cardiaca_max',
    'angina_ejercicio', 'depresion_st', 'pendiente_st',
    'vasos_principales', 'thal', 'enfermedad_cardiaca'
]

# Exploración de calidad del dato
print("--- Tipos de datos ---")
print(df.dtypes)

print("\n--- Valores nulos por columna ---")
print(df.isnull().sum())

print("\n--- Estadísticas generales ---")
df.describe()
```

```python
# Configuración general de gráficas
plt.rcParams['figure.figsize'] = (10, 5)
sns.set_style("whitegrid")

# --- Gráfica 1: Distribución de edades ---
plt.figure()
sns.histplot(df['edad'], bins=20, color='steelblue', kde=True)
plt.title('Distribución de Edades de Pacientes')
plt.xlabel('Edad')
plt.ylabel('Cantidad de Pacientes')
plt.tight_layout()
plt.savefig('grafica_edades.png', dpi=150)
plt.show()

# --- Gráfica 2: Pacientes con y sin enfermedad cardíaca ---
plt.figure()
df['enfermedad_cardiaca'].value_counts().plot(
    kind='bar',
    color=['steelblue', 'tomato'],
    edgecolor='white'
)
plt.title('Pacientes con y sin Enfermedad Cardíaca')
plt.xlabel('0 = Sin enfermedad | 1 = Con enfermedad')
plt.ylabel('Cantidad de Pacientes')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('grafica_enfermedad.png', dpi=150)
plt.show()

print("✅ Gráficas generadas correctamente")
```

```python
# --- Gráfica 3: Distribución por sexo ---
plt.figure()
sexo_labels = {0: 'Femenino', 1: 'Masculino'}
df['sexo_texto'] = df['sexo'].map(sexo_labels)
sns.countplot(x='sexo_texto', data=df, palette=['tomato', 'steelblue'])
plt.title('Distribución de Pacientes por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Cantidad de Pacientes')
plt.tight_layout()
plt.savefig('grafica_sexo.png', dpi=150)
plt.show()

# --- Gráfica 4: Colesterol promedio por grupo de edad ---
plt.figure()
df['grupo_edad'] = pd.cut(
    df['edad'],
    bins=[29, 40, 50, 60, 77],
    labels=['29-40', '41-50', '51-60', '61-77']
)
colesterol_edad = df.groupby('grupo_edad', observed=True)['colesterol'].mean().reset_index()
sns.barplot(x='grupo_edad', y='colesterol', data=colesterol_edad, palette='Blues_d')
plt.title('Colesterol Promedio por Grupo de Edad')
plt.xlabel('Grupo de Edad')
plt.ylabel('Colesterol Promedio (mg/dl)')
plt.tight_layout()
plt.savefig('grafica_colesterol.png', dpi=150)
plt.show()

print("✅ Gráficas generadas correctamente")
```

```python
# --- Gráfica 5 corregida ---
plt.figure()
sns.boxplot(
    x='sexo_texto',
    y='frec_cardiaca_max',
    hue='enfermedad_cardiaca',
    data=df,
    palette={0: 'steelblue', 1: 'tomato'}
)
plt.title('Frecuencia Cardíaca Máxima por Sexo y Diagnóstico')
plt.xlabel('Sexo')
plt.ylabel('Frecuencia Cardíaca Máxima')
plt.legend(title='Enfermedad', labels=['Sin enfermedad', 'Con enfermedad'])
plt.tight_layout()
plt.savefig('grafica_frecuencia.png', dpi=150)
plt.show()

# --- Exportar datos limpios para Power BI ---
df_export = df.copy()

# Convertir variables numéricas a texto descriptivo para Power BI
df_export['sexo'] = df_export['sexo'].map({0: 'Femenino', 1: 'Masculino'})
df_export['enfermedad_cardiaca'] = df_export['enfermedad_cardiaca'].map({0: 'Sin enfermedad', 1: 'Con enfermedad'})
df_export['angina_ejercicio'] = df_export['angina_ejercicio'].map({0: 'No', 1: 'Sí'})

# Eliminar columnas auxiliares
df_export = df_export.drop(columns=['sexo_texto', 'grupo_edad'])

# Guardar CSV
df_export.to_csv('pacientes_limpios.csv', index=False)

print("✅ Análisis completo")
print(f"📁 Archivo exportado: pacientes_limpios.csv")
print(f"📊 Total registros exportados: {len(df_export)}")
```
